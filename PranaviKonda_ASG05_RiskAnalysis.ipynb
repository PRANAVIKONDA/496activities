```python
import seaborn as sns
import pandas as pd
import numpy as np
riskDF= pd.DataFrame(columns=['risk', 'likelihood','impact','score'])
riskDF['risk']=['Employee Disbelief','Outside Breach','Unpatched devices','Internal attacks','Shadow IT','Own Device Security']
riskDF['likelihood']=[1.1,0.9,0.7,0.5,0.3,0.1]
riskDF['impact']=[36,32,14,11,4,3]
riskDF['score']=riskDF['impact']*riskDF['likelihood']
print(riskDF)
scoresPivot=riskDF.pivot('impact','likelihood','score')
print(scoresPivot)
labelsPivot=riskDF.pivot('impact','likelihood','risk')
print(labelsPivot)
sns.set(rc={'figure.figsize':(16,16)})
p1=sns.heatmap(scoresPivot, cmap='coolwarm', annot=labelsPivot, fmt='')
```
