```python
# TRIANGULAR
import numpy as np
import matplotlib.pyplot as plt

triangularSample=np.random.triangular(2,7, 11, size=5000)
plt.hist(triangularSample, bins=200, density=True)
plt.show()
```

```python
#TENSOR FLOW
import tensorflow_probability as tfd
tfd = tfd.distributions

dist= tfd.PERT(low=1., peak=7., high=11., temperature=4.)

print(dist.sample(10))

print(dist.prob(6.))
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

Opt=4
Ml=7
Pess=14


plt.figsize=(5,5)
for Temp in range(5,8):
  dist=tfd.PERT(low=Opt, peak=Ml, high=Pess, temperature=Temp)
  estimates=dist.sample(2000)
  ax=sns.distplot(estimates)

ax.legend(range(5,8))
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

Opt=4
Ml=7
Pess=14

fig, axs=plt.subplots(2,2, figsize=(5,5))
i=0
j=0
for Temp in range(8,12):
  dist=tfd.PERT(low=Opt, peak=Ml, high=Pess, temperature=Temp)
  estimates=dist.sample(5000)
  axs[i//2,i%2].hist(estimates, bins=20, density=True)
  i+=1

plt.tight_layout()
plt.show()
```
